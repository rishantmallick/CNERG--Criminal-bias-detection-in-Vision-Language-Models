```python
from bert_score import score
import csv
# Create CSV once with no header (append mode)
csv_file = "results1.csv"
from rouge_score import rouge_scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'])
for i in range(0,20):
  target_text = [f"{df.iloc[i]['target_text']}"]
  summaries = [f"{df.iloc[i]['summaries']}"]
  original_text = f"{df.iloc[i]['target_text']}"
  summary = f"{df.iloc[i]['summaries']}"
  P,R,F1 = score(target_text,summaries,lang="en", model_type="bert-base-uncased")
  scores = scorer.score(summary, original_text)
  print(f"Paper {i+1}")
  print("Precision:", P.mean().item())
  print("Recall:", R.mean().item())
  print("F1:", F1.mean().item())
  print("Content:",scores)
  paper_id = i+1
  precision = P.mean().item()
  recall = R.mean().item()
  f1 = F1.mean().item()
  content_scores = scores

  # Convert to a single string exactly how you want
  row_text = (
    f"Paper {paper_id} "
    f"Precision: {precision} "
    f"Recall: {recall} "
    f"F1: {f1} "
    f"Content: {content_scores}"
  )
  with open(csv_file, "a", newline="") as f:
        writer = csv.writer(f)
        writer.writerow([row_text])
```

```python
import pandas as pd
import re
file_path = "/content/results of the research paper.csv"
df = pd.read_csv(file_path)
column_name = df.columns[0]  # Get the first column name
meanf1 = []
print(f"\nColumn name to parse: '{column_name}'")

# Extract overall F1 score from the column name
overall_f1_match = re.search(r'F1:\s*@?([\d.]+)', column_name)
if overall_f1_match:
    overall_f1 = overall_f1_match.group(1)
    print(f"Overall F1 Score: {overall_f1}")

# Extract ROUGE F1 scores from the column name
rouge_f1_matches = re.findall(r'fmeasure=([\d.]+)', column_name)
if rouge_f1_matches and len(rouge_f1_matches) >= 3:
    print(f"\nROUGE F1 Scores from column name:")
    print(f"ROUGE-1: {rouge_f1_matches[0]}")
    print(f"ROUGE-2: {rouge_f1_matches[1]}")
    print(f"ROUGE-L: {rouge_f1_matches[2]}")
meanf1.append(float(overall_f1))
```
